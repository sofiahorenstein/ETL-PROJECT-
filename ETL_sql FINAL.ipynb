```python
#JOINED TABLE IN SQL
sql = """
SELECT s.unnamed_iowa, s.invoice_itemnumber, s.date, s.city, s.zip_code, s.county_number, s.county, s.brand, s.bottles_sold, s.sale, s.year, s.month AS iowa_liquor_sale, r.unnamed_review, r.id, r.reviews_date, r.reviews_rating, r.reviews_sourceurls, r.reviews_text, r.reviews_title, r.reviews_username, r.sourceurls AS iowa_liquor_review
FROM iowa_liquor_sale s
INNER JOIN iowa_liquor_review r ON r.brand = s.brand
"""
joined_df = pd.read_sql_query(sql, engine)
joined_df


```

# DATA ANALYSIS 

```python
#mean values for each unique brand
sql = """
SELECT brand, AVG(reviews_rating)
FROM liquor_brand_merged
GROUP BY brand;
"""
brand_rating = pd.read_sql_query(sql, engine)
brand_rating
```

```python
#total sales by brand 
sql = """
SELECT brand, SUM(sale)
FROM liquor_brand_merged
GROUP BY brand;
"""
total_brand_sales = pd.read_sql_query(sql, engine)
total_brand_sales
```

```python
#total purchase count by brand 
sql = """
SELECT brand, COUNT(sale)
FROM liquor_brand_merged
GROUP BY brand;
"""
total_brand_purchases = pd.read_sql_query(sql, engine)
total_brand_purchases
```

```python
#total item sales by county (highest)
sql = """
SELECT SUM(sale), county
FROM liquor_brand_merged
GROUP BY county
ORDER BY sum(sale) DESC;;
"""
total_county_sales = pd.read_sql_query(sql, engine)
total_county_sales.head()
```

```python
#total purchase count by county (highest)
sql = """
SELECT COUNT(sale), county
FROM liquor_brand_merged
GROUP BY county
ORDER BY COUNT(sale) DESC;;
"""
total_county_purchases = pd.read_sql_query(sql, engine)
total_county_purchases.head()
```

```python
#total item sales by county (lowest)
sql = """
SELECT SUM(sale), county
FROM liquor_brand_merged
GROUP BY county
ORDER BY sum(sale) ASC;;
"""
total_county_sales = pd.read_sql_query(sql, engine)
total_county_sales.head()
```

```python
#total purchase count by county (lowest)
sql = """
SELECT COUNT(sale), county
FROM liquor_brand_merged
GROUP BY county
ORDER BY sum(sale) ASC;;
"""
total_county_sales = pd.read_sql_query(sql, engine)
total_county_sales.head()
```
